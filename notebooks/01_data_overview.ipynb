## Summary — Raw Traffy Data Overview

### 1) Dataset Snapshot
- File loaded: `../data/bangkok_traffy.csv`
- Shape: **787,026 rows**, **16 columns**
- Columns:
  - `ticket_id`, `type`, `organization`, `comment`
  - `photo`, `photo_after`
  - `coords`, `address`, `subdistrict`, `district`, `province`
  - `timestamp`, `state`, `star`, `count_reopen`, `last_activity`

### 2) Column Types and Structure
- `df.info()` shows:
  - 14 columns are `object` (string-like)
  - 1 column is `float64` → `star`
  - 1 column is `int64` → `count_reopen`
- `timestamp` and `last_activity` are initially `object` and later converted to datetime
- Memory usage ~96 MB

### 3) Missing Values Overview
Top missing columns (from `missing` table):
- `star`: **512,929 missing** (~65.17%)
- `photo_after`: **145,717 missing** (~18.51%)
- `ticket_id`, `comment`, `address`: **8,772 missing each** (~1.11%)
- `organization`: 571 missing (~0.07%)
- `subdistrict`: 566 missing (~0.07%)
- `district`: 561 missing (~0.07%)
- `province`: 195 missing (~0.02%)
- `photo`: 115 missing (~0.01%)
- `type`: 97 missing (~0.01%)
- No missing values in:
  - `coords`, `timestamp`, `state`, `count_reopen`, `last_activity`

### 4) Uniqueness and Categorical Complexity
From the unique counts:
- `type`: **17,469** unique values  
  → very high cardinality, many combinations like `{คลอง,เสียงรบกวน,ถนน}` → must standardize later
- `province`: 81 unique
- `district`: 144 unique
- `subdistrict`: 372 unique
- `organization`: **70,208** unique (very high, likely too granular for direct modeling)
- `state`: 3 unique values:
  - `'เสร็จสิ้น'`
  - `'กำลังดำเนินการ'`
  - `'รอรับเรื่อง'`
